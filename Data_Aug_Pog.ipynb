```python
def zoom(img_path, zoom_range_list=[.9, 1.1]):
    img= load_img(img_path)
    data = img_to_array(img)
    samples = expand_dims(data, 0)
    datagen = ImageDataGenerator(zoom_range=zoom_range_list)
    it = datagen.flow(samples, batch_size=1)
    batch=it.next()
    image=batch[0].astype('uint8')
    pyplot.imshow(image)
    pyplot.show()
```

```python
def h_flip(img_path):
    img = load_img(img_path)
    data = img_to_array(img)
    samples = expand_dims(data, 0)
    datagen = ImageDataGenerator(horizontal_flip=True)
    it = datagen.flow(samples, batch_size=1)
    batch = it.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    # plot raw pixel data
    pyplot.imshow(image)
    # show the figure
    pyplot.show()
```

```python
def rotate(img_path, rot_range=360):
    img= load_img(img_path)
    img = img_to_array(img)
    samples=expand_dims(img,0)
    datagen=ImageDataGenerator(rotation_range=rot_range)
    it=datagen.flow(samples, batch_size=1)
    batch=it.next()
    image=batch[0].astype('uint8')
    pyplot.imshow(image)
```

```python
def shear(img_path):
    img = load_img(img_path)
    rand_var = np.random.uniform()/5
    afine_tf = tf.AffineTransform(shear=rand_var)
    modified = tf.warp(img, inverse_map=afine_tf)
    io.imshow(modified)
    plt.show()
```
